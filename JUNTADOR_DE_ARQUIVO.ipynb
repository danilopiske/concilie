```python
import pandas as pd
import numpy as np
import glob
import os

# Definir diretamente os diretórios das pastas
pasta_df = r"D:\Arquivos_Concilie\Nova pasta\Ec 1051121873\lancamentos"
pasta_dv = r"D:\Arquivos_Concilie\Nova pasta\Ec 1051121873\Vendas"

# Leitura automática dos arquivos df (header=2)
arquivos_df = sorted(glob.glob(os.path.join(pasta_df, '*.xlsx')))
print(f'Total de arquivos df encontrados: {len(arquivos_df)}')
lista_df = []
for idx, arquivo in enumerate(arquivos_df, start=1):
    print(f'Lendo arquivo df {idx}/{len(arquivos_df)}: {os.path.basename(arquivo)}')
    temp_df = pd.read_excel(arquivo, header=2)
    lista_df.append(temp_df)
df = pd.concat(lista_df, ignore_index=True)
print('Concatenação dos df finalizada.')

# Leitura automática dos arquivos dv (header=9)
arquivos_dv = sorted(glob.glob(os.path.join(pasta_dv, '*.xlsx')))
print(f'Total de arquivos dv encontrados: {len(arquivos_dv)}')
lista_dv = []
for idx, arquivo in enumerate(arquivos_dv, start=1):
    print(f'Lendo arquivo dv {idx}/{len(arquivos_dv)}: {os.path.basename(arquivo)}')
    temp_dv = pd.read_excel(arquivo, header=9)
    lista_dv.append(temp_dv)
dv = pd.concat(lista_dv, ignore_index=True)
print('Concatenação dos dv finalizada.')

# Garantir que data_transacao está em datetime (formato brasileiro dd/mm/yyyy)
df['Data da Transação'] = pd.to_datetime(df['Data da Transação'], dayfirst=True, errors='coerce')
dv['Data da venda'] = pd.to_datetime(dv['Data da venda'], dayfirst=True, errors='coerce')

df['ano'] = df['Data da Transação'].dt.year
dv['ano'] = dv['Data da venda'].dt.year

df_anos = df['ano'].unique()
dv_anos = dv['ano'].unique()

# Para cada ano, criar arquivo Excel com duas abas: dados filtrados e mínimos agrupados
for ano in df_anos:
    df_ano = df[df['ano'] == ano]
    minimos = df_ano.groupby(['Produto cielo'])['Percentual de Desconto'].min().reset_index()
    with pd.ExcelWriter(fr'D:\Vendas BATEL - 1051121873\Vendas\lancamentos_1051121873_{ano}.xlsx') as writer:
        df_ano.to_excel(writer, sheet_name='dados', index=False)
        minimos.to_excel(writer, sheet_name='minimos', index=False)

for ano in dv_anos:
    dv_ano = dv[dv['ano'] == ano]
    minimos = dv_ano.groupby(['Forma de pagamento', 'Bandeira'])['Taxa administrativa (MDR)'].min().reset_index()
    with pd.ExcelWriter(fr'D:\Vendas BATEL - 1051121873\Vendas\vendas_1051121873_{ano}.xlsx') as writer:
        dv_ano.to_excel(writer, sheet_name='dados', index=False)
        minimos.to_excel(writer, sheet_name='minimos', index=False)

print('✅ Processamento concluído!')
```
